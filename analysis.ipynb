#### Where should we deploy a food access program?
- Where can we have the most impact? Affect the most people
- Where is there a worsening problem? (Did some EDA to discover that a lot of places have a lot of people that have low access but less people were affected in 2010 vs 2015)

```python
access_df['LACCESS_PRC_TOTAL15'] = (access_df['LACCESS_POP15']/access_df['LACCESS_POP15'].sum())*100  # percentage of total
access_df['LACCESS_PRC_TOTAL10'] = (access_df['LACCESS_POP10']/access_df['LACCESS_POP10'].sum())*100  # percantage of total
access_df['PCH_LACCESS_PRC_TOTAL_10_15'] = access_df['LACCESS_PRC_TOTAL15'] - access_df['LACCESS_PRC_TOTAL10']  # difference in percentage from 2015 to 2010
access_df['CH_LACCESS_TOTAL_10_15'] = access_df['LACCESS_POP15'] - access_df['LACCESS_POP10']  # difference in population from 2015 to 2010
access_df['County_State'] = access_df["County"] + ', ' + access_df["State"]  # for x-axis labels
```

```python
# get top 5 difference in percentage from 2015 to 2010, keeping more columns to try a few different viz options

top_5 = access_df.sort_values(by=[
    'PCH_LACCESS_PRC_TOTAL_10_15'
], ascending=False)[[
    'FIPS', 'County_State', 
    'PCH_LACCESS_PRC_TOTAL_10_15', 
    'CH_LACCESS_TOTAL_10_15',
    'LACCESS_PRC_TOTAL15',
    'LACCESS_PRC_TOTAL10',
    'LACCESS_POP15', 'LACCESS_POP10'
]].head()
```

##### Trying different bar charts. A lot of categorical data, want this to be understandable to non-technical people

#### Which subgroup of the population might benefit the most from the program?
- subgroup benefitting the most (probably households because that's what we're targeting) but a breakdown in Clark, NV shows these are the demographics making up larger chunks (obviously they aren't mutually exclusive)

```python
# subgroups
location_variables = ['FIPS', 'County_State']

demographic_variables = variable_list_df.loc[
    (variable_list_df['Subcategory Name']=='Demographics') &
    (variable_list_df['Variable Code'].isin(access_df.columns))
]['Variable Code'].tolist()
```

```python
# drop variables from 2010, PCT, and PCH variables
print(len(demographic_variables))
for var in set(demographic_variables):
    if var[:3] == 'PCT':
        demographic_variables.remove(var)
    elif var[:3] == 'PCH':
        demographic_variables.remove(var)
    elif var[-2:] == '10':
        demographic_variables.remove(var)
    else:
        continue
print(len(demographic_variables))

relevant_variables = location_variables + demographic_variables

Clark_NV_demographic = access_df[relevant_variables].loc[access_df['County_State'] == 'Clark, NV']
Clark_NV_demographic[demographic_variables] = Clark_NV_demographic[demographic_variables].astype('Float64')
```

```python
# transpose row and order demographic variables so bar char will look pretty
ordered_demographic = Clark_NV_demographic[demographic_variables].T.sort_values(by=1748, ascending=False).index.tolist()
```

```python
subgroup_labels = [
    'White',
    'Hispanic',
    'Children',
    'Multiracial',
    'Seniors',
    'Black',
    'Asian',
    'American Indian or Alaskan Native',
    'Hawaiian or Pacific Islander'
]
```

```python
ax2 = Clark_NV_demographic.plot(kind='bar', x='County_State', y=ordered_demographic, colormap='Dark2', figsize=(12, 9), width=1.0)
ax2.legend(subgroup_labels, loc='best', fontsize='large')
ax2.set(xlabel='', ylabel='Number of Individuals')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=0)
ax2.margins(x=0, y=0.1)
```

##### figuring out joins to city data (did some EDA on this as well)
- five hundred is place level and census tract
- access is county level
- access county codes need a zero at the front if 4 digits (county codes are 5, dataset dropped zeros at the front for some reason)
- five hundred needs to split on place_tractID hyphen because the first five digits of census tract code are county codes with zeros

```python
five_hun_df['Place_TractID'].isna().sum()  # checking for null values
```

```python
five_hun_df['Place_TractID'] = five_hun_df['Place_TractID'].astype(str)  # convert to string
five_hun_df[['CityFIPS', 'CountyFIPS']] = five_hun_df['Place_TractID'].str.split(pat='-', n=0, expand=True)  # split on hyphen
```

```python
five_hun_df['CountyFIPS'].isna().sum()  # noticed there were nulls in the new column
```

```python
five_hun_df.loc[five_hun_df['CountyFIPS'].isna()]['PlaceFIPS'].value_counts()  # all from the same place
```

```python
five_hun_df.loc[five_hun_df['PlaceFIPS'] == 15003]  # it is Honolulu, cool!
```

```python
# filling null counties with 15003 since the place and county code are the same here
five_hun_df['CountyFIPS'] = five_hun_df['CountyFIPS'].fillna('15003')
five_hun_df['CountyFIPS'].isna().sum()  # verify it worked
```

```python
five_hun_df['FIPS'] = five_hun_df['CountyFIPS'].apply(lambda x: x[:5] if len(x)>5 else x)  # trimming the extra stuff after county code
five_hun_df = five_hun_df.drop(columns='CountyFIPS')  # remove CountyFIPS, names need to be the same for join bc pandas
```

```python
access_df['FIPS'] = access_df['FIPS'].astype(str).apply(lambda x: f'0{x}' if len(x) == 4 else x)  # add zeros to access FIPS so we can join
```

```python
access_city_df = five_hun_df.merge(access_df, on='FIPS')
```

##### time to find some variables that we think are correlated to food insecurity/low grocery store access
- HIGHCHOL_CrudePrev, Model-based estimate for crude prevalence of high cholesterol among adults aged >=18 years who have been screened in the past 5 years, 2017
- CANCER_CrudePrev, Model-based estimate for crude prevalence of cancer (excluding skin cancer) among adults aged >=18 years, 2017
- BPHIGH_CrudePrev, Model-based estimate for crude prevalence of taking medicine for high blood pressure control among adults aged >=18 years with high blood pressure, 2017
- CHD_CrudePrev, Model-based estimate for crude prevalence of coronary heart disease among adults aged >=18 years, 2017
- Crude Prevalence definition https://www.cdc.gov/cancer/uscs/about/hints.htm
- total number of cases or deaths divided by the total population and multiplied by 100,000
- if I had more time I'd deconstruct these back to the actual population estimates for improved interpretation
- find average of each in Clark County
- compare to averages available in complete dataset?

```python
clark_impact = access_city_df[[
    'County_State', 'PlaceName', 'HIGHCHOL_CrudePrev', 'CANCER_CrudePrev', 'BPHIGH_CrudePrev', 'CHD_CrudePrev'
]].loc[access_city_df['County_State'] == 'Clark, NV']
```

```python
print(clark_impact['HIGHCHOL_CrudePrev'].mean())
print(clark_impact['HIGHCHOL_CrudePrev'].median())  # for outlier safety

print(clark_impact['CANCER_CrudePrev'].mean())
print(clark_impact['CANCER_CrudePrev'].median())  # for outlier safety

print(clark_impact['BPHIGH_CrudePrev'].mean())
print(clark_impact['BPHIGH_CrudePrev'].median())  # for outlier safety

print(clark_impact['CHD_CrudePrev'].mean())
print(clark_impact['CHD_CrudePrev'].median())  # for outlier safety
```

```python
print(access_city_df['HIGHCHOL_CrudePrev'].mean())
print(access_city_df['HIGHCHOL_CrudePrev'].median())  # for outlier safety

print(access_city_df['CANCER_CrudePrev'].mean())
print(access_city_df['CANCER_CrudePrev'].median())  # for outlier safety

print(access_city_df['BPHIGH_CrudePrev'].mean())
print(access_city_df['BPHIGH_CrudePrev'].median())  # for outlier safety

print(access_city_df['CHD_CrudePrev'].mean())
print(access_city_df['CHD_CrudePrev'].median())  # for outlier safety
```

```python
data = {
    'Type': ['Clark, NV Average', 'Dataset Average'],
    'High Cholesterol': [clark_impact['HIGHCHOL_CrudePrev'].mean(), access_city_df['HIGHCHOL_CrudePrev'].mean()],
    'Cancer': [clark_impact['CANCER_CrudePrev'].mean(), access_city_df['CANCER_CrudePrev'].mean()],
    'High Blood Pressure': [clark_impact['BPHIGH_CrudePrev'].mean(), access_city_df['BPHIGH_CrudePrev'].mean()],
    'Coronary Heart Disease': [clark_impact['CHD_CrudePrev'].mean(), access_city_df['CHD_CrudePrev'].mean()]
}

averages_df = pd.DataFrame.from_dict(data)


# sort here to put Clark over Dataset
ax3 = averages_df.sort_values(by='Type', ascending=False).plot(
    kind='barh', y=['Coronary Heart Disease', 'Cancer', 'High Blood Pressure', 'High Cholesterol'], 
    x='Type', colormap='Dark2'
)

ax3.set(xlabel='Model Estimated Crude Prevelance', ylabel='')
for p in ax3.containers:
    ax3.bar_label(p, label_type='edge', fmt =f'%.02f', padding=4, fontsize='large')
ax3.legend(reverse=True)
```
